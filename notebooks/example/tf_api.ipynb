```python
# Import tensorflow
import tensorflow as tf
```

```python
sess = tf.Session()
```

```python
# Constants, Sequences, and Random Values
# link: https://www.tensorflow.org/versions/r0.10/api_docs/python/constant_op/
```

```python
x = tf.linspace(-3.0, 3.0, 32)
print(sess.run(x))
print(sess.run(tf.multinomial(tf.log([[10., 10., 10., 9., 1.],[10., 10., 1., 2., 5.]]), 5)))
```

```python
print(sess.run(tf.multiply(2, 3, name=None)))
```

```python
t_1 = ['apple', 'peach', 'banana']
print(sess.run(tf.zeros_like(t_1)) )
```
