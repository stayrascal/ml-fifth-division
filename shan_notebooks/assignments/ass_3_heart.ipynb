```python
import tensorflow as tf
import pandas as pd
import numpy as np
import random
```

```python
data_file = '../data/heart.csv'
origin_data = pd.read_csv(data_file)
origin_data
```

```python
def load_data(data):
    positive_example = data[data['chd'] == 1]
    negative_example = data[data['chd'] == 0]
    
    positive_msk = np.random.rand(len(positive_example)) < 0.8
    negative_msk = np.random.rand(len(negative_example)) < 0.8
#    print('len(positive_example)\n', len(positive_example))
 #   print('len(negative_example)\n', len(negative_example))
 #   print('len(positive_example[positive_msk])\n)', len(positive_example[positive_msk]))
 #   print('len(negative_example[negative_msk])\n)', len(negative_example[negative_msk]))
 #   print('len(positive_example[~positive_msk])\n)', len(positive_example[~positive_msk]))
 #   print('len(negative_example[~negative_msk])\n)', len(negative_example[~negative_msk]))
    
    train_dataset = pd.concat([positive_example[positive_msk], negative_example[negative_msk]])
    test_dataset = pd.concat([positive_example[~positive_msk], negative_example[~negative_msk]])
#    print('len(train_dataset)\n', len(train_dataset))
#    print('len(test_dataset)\n', len(test_dataset))
    return train_dataset, test_dataset
    
#axis : {0 or ‘index’, 1 or ‘columns’}, default 0
#0 or ‘index’: apply function to each column
#1 or ‘columns’: apply function to each row
def normailize_feature(dataset):
    return dataset.apply(lambda x: (x-np.min(x))/(np.max(x)-np.min(x)), axis=0)

def transform_famhist(dataset):
    tmp = dataset.replace({'famhist': {'Present': 1, 'Absent': 0}})
  #  print('normailize_feature(tmp).values', normailize_feature(tmp).values)
    return normailize_feature(tmp).values

def randomize(dataset, labels):
    permutation = np.random.permutation(labels.shape[0])
    shuffled_dataset = dataset[permutation]
    shuffled_labels = labels[permutation]
    return shuffled_dataset, shuffled_labels

def to_one_hotting(data, num_lables=1):
    return (np.arange(num_lables) == data[:,None]).astype(np.float32)

def generate_data(dataset):
    data = dataset.iloc[:,0:9]
    print('len(data)\n', len(data))
    labels = dataset['chd']
    print('labels.values\n', labels.values)
    print('labels\n',labels)
    return randomize(transform_famhist(data), to_one_hotting(labels.values))


train_data, test_data = load_data(origin_data)
train_feature, train_labels = generate_data(train_data)
test_feature, test_labels = generate_data(test_data)

print(train_feature.shape, train_labels.shape)
print(test_feature.shape, test_labels.shape)
print('test_labels \n', test_labels)
```

```python
learning_rate = 0.01
training_echos = 100
batch = 10
```

```python
X = tf.placeholder(dtype = np.float32, shape = [None, 9], name='X')
Y = tf.placeholder(dtype = np.float32, shape = [None, 1], name='Y')

W = tf.Variable(tf.random_normal([9, 1]), name='W')
b = tf.Variable(tf.zeros([1]), name='b')

logits = tf.matmul(X, W) + b
entropy = tf.nn.softmax_cross_entropy_with_logits(labels = Y, logits = logits)
loss = tf.reduce_mean(entropy)

optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)

pred = tf.nn.softmax(logits)
# 测试模型
correct_prediction = tf.equal(tf.argmax(pred, 1), tf.argmax(Y, 1))
# 计算准确度
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

# 初始化所有变量
init = tf.initialize_all_variables()
```

```python
with tf.Session() as sess:
    sess.run(init)
    n_batches = int(len(train_feature)/batch)
    print(n_batches)
  #  for n in range (training_echos):
    total_loss=0
    for index in range (n_batches):
        X_batch = train_feature[index*batch:(index+1)*batch]
        print('X_batch \n', X_batch)
        Y_batch = train_labels[index*batch:(index+1)*batch]
        print('Y_batch \n',Y_batch)    
        _, loss = sess.run([optimizer, loss], feed_dict={X: X_batch, 
                                                         Y: Y_batch})
        total_loss += loss
        print('total_loss \n',total_loss)
        
    test_accuracy = sess.run(accuracy, feed_dict={X: test_feature, Y: test_labels})
       # accuracy = test_accuracy / len(test_labels)
    print('Average loss:{0} at epoch:{1}'.format(total_loss/batch, i), ' and accuracy is :', test_accuracy)
        
print('Optimization Finished!')
avg_accuracy = sess.run(accuracy, feed_dict={X: test_feature, Y: test_labels})
print('Final Accuracy is :', avg_accuracy)
        
```
